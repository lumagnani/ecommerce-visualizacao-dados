```python
import matplotlib.pyplot as plt #criação de gráficos
import pandas as pd #leitura e análise de dados
import seaborn as sns #gráficos mais complexos
import numpy as np #auxilia com cálculos
```

```python
#abre a planilha e transforma em data frame
df = pd.read_csv('ecommerce_preparados.csv') 

#mostra a quantidade total de linhas e colunas dentro do Data Frame
print(df.info()) 
```

```python
#Mostra a média, mínimo, máximo, etc.
df.describe()
```

## Distribuição das Notas dos Produtos (Histograma)
### As notas dos produtos são, em sua maioria, altas ou baixas?


##### O histograma mostra que a maioria dos produtos possui notas concentradas nas faixas mais altas da escala, indicando avaliações predominantemente positivas. Observa-se uma menor quantidade de produtos com notas baixas, o que sugere um bom nível de satisfação dos consumidores em relação aos produtos disponíveis na plataforma.


```python
plt.hist(df['Nota'], bins=10) #divide o gráfico em 10 barras
plt.title('Distribuição das Notas dos Produtos')
plt.xlabel('Nota')
plt.ylabel('Quantidade de Produtos')
plt.show()
```

## 2. Relação entre Preço e Avaliação dos Produtos (Gráfico de Dispersão)

### O preço influencia na nota?


##### O gráfico de dispersão indica que não existe uma relação clara e direta entre o preço dos produtos e suas notas. Produtos com preços mais baixos e mais altos apresentam avaliações semelhantes, sugerindo que o preço não é um fator determinante para a satisfação do consumidor.

```python
plt.scatter(df['Preço'], df['Nota'])
plt.title('Preço vs Nota')
plt.xlabel('Preço')
plt.ylabel('Nota')
plt.show()
```

## 3.Correlação entre Variáveis Numéricas do E-commerce (Mapa de Calor)
### Quais variáveis apresentam maior relação entre si?

##### O mapa de calor permite identificar o grau de correlação entre as variáveis numéricas do conjunto de dados. Observa-se uma correlação mais forte entre algumas variáveis relacionadas, enquanto outras apresentam correlação fraca ou inexistente, indicando independência entre determinados atributos.

```python
correlacao = df[['Preço', 'Nota', 'N_Avaliações', 'Desconto']].corr()
plt.imshow(correlacao)
plt.colorbar()
plt.xticks(range(4), correlacao.columns)
plt.yticks(range(4), correlacao.columns)
plt.title('Correlação entre variáveis')
plt.show()
```

## 4. Frequência de Produtos por Marca (Gráfico de Barras)

### Quais marcas aparecem com maior frequência na base de dados?

##### O gráfico de barras mostra que algumas marcas aparecem com maior frequência na base de dados, indicando maior presença ou variedade de produtos dessas marcas no e-commerce. Outras marcas possuem participação menor, o que pode indicar menor oferta ou menor relevância no conjunto analisado.


```python
#Barras
##Quais marcas aparecem mais?

df['Marca'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Marcas')
plt.xlabel('Marca')
plt.ylabel('Quantidade')
plt.show()
```

## 5. Distribuição dos Produtos por Gênero (Gráfico de Pizza)
### Como os produtos estão distribuídos de acordo com o gênero?

##### O gráfico de pizza evidencia a distribuição percentual dos produtos por gênero. Nota-se que alguns gêneros concentram a maior parte dos produtos, enquanto outros representam parcelas menores, o que ajuda a compreender o foco do catálogo do e-commerce.

```python
genero_counts = df['Gênero'].value_counts()

plt.figure(figsize=(8, 8))

plt.pie(
    genero_counts,
    autopct='%1.1f%%',
    startangle=90
)

plt.legend(
    genero_counts.index,
    title='Gênero',
    loc='center left',
    bbox_to_anchor=(1, 0.5)
)

plt.title('Distribuição por Gênero')
plt.show()
```

## 6. Concentração dos Preços dos Produtos (Gráfico de Densidade)
### Em quais faixas de preço os produtos estão mais concentrados?

##### O gráfico de densidade mostra que os preços dos produtos se concentram na faixa de 0 a 200 reais, indicando que a maioria dos itens está posicionada entre esses valores. Isso sugere uma estratégia de precificação focada em determinados segmentos de mercado.

```python
df['Preço'].plot(kind='density')
plt.title('Densidade dos Preços')
plt.xlabel('Preço')
plt.show()
```

## 7. Tendência entre Preço e Nota dos Produtos (Regressão Linear)
### Existe uma tendência entre o preço do produto e sua avaliação?

##### A linha de regressão indica que não há uma tendência forte entre o preço dos produtos e suas notas. A inclinação suave da linha sugere que variações no preço não impactam significativamente a avaliação dos consumidores.

```python
df_reg = df[['Preço', 'Nota']].dropna()
x = df_reg['Preço']
y = df_reg['Nota']


coef = np.polyfit(x, y, 1)
linha = np.poly1d(coef)

plt.scatter(x, y, alpha=0.3)
plt.plot(x, linha(x))
plt.xlim(0, 300)
plt.title('Regressão Linear: Preço vs Nota')
plt.xlabel('Preço')
plt.ylabel('Nota')
plt.show()
```
